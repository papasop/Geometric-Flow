# CIFAR-10 Geometric Flow Benchmark

Run the first cell to clone the AGI repository, install dependencies, download CIFAR-10, and compare Adam, pure geometric, and hybrid optimizer modes.

```python
# @title Run CIFAR-10 Benchmark
# @markdown This runs Adam, pure geometric, and hybrid modes with warm-up steps 10, 30, 50, and 80.

%%bash
set -e
rm -rf /content/AGI
git clone https://github.com/papasop/AGI.git /content/AGI
cd /content/AGI
pip install -e . torchvision > /dev/null
python experiments/run_cifar10_benchmark.py \
  --download \
  --steps 500 \
  --trials 3 \
  --hybrid-warmup-steps "10,30,50,80" \
  --out artifacts/cifar10_benchmark_results.csv
echo
echo "Results CSV:"
cat artifacts/cifar10_benchmark_results.csv
```

```python
# @title Optional: Fast Synthetic Milestone Smoke Test
# @markdown Use this when you only want a quick functional check without downloading CIFAR-10.

%%bash
set -e
cd /content/AGI
python experiments/train_cifar10_geo.py \
  --dataset synthetic \
  --mode all \
  --trials 1 \
  --steps 50 \
  --adam-warmup-steps 48 \
  --seed 32 \
  --train-samples 256 \
  --eval-samples 128 \
  --batch-size 32 \
  --channels 16 \
  --use-fisher \
  --preconditioner diagonal \
  --precond-scale 0.5 \
  --max-grad-norm 2.0 \
  --grad-smoothing 0.0 \
  --out artifacts/synthetic_cifar_milestone.csv
cat artifacts/synthetic_cifar_milestone.csv
```
